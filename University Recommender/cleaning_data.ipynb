```python
def custom_title_case(string):
    # Split the string into words
    words = string.split()
    
    # List of words to exclude from capitalization
    excluded_words = ["to", "of"]
    
    # Capitalize the first letter of each word except excluded words
    capitalized_words = [word.capitalize() if word.lower() not in excluded_words else word.lower() for word in words]
    
    # Join the words back into a string
    return ' '.join(capitalized_words)
```

```python
university["University"] = university["University"].apply(custom_title_case)
university.head()
```

```python
def split_string_with_exceptions(series):
    # Split each string in the Series based on provided keywords but excluding the keywords themselves
    email_part = series.str.split("Email:").str[1].str.split("Fax:").str[0].str.strip()
    fax_part = series.str.split("Fax:").str[1].str.split("URL:").str[0].str.strip()
    url_part = series.str.split("URL:").str[1].str.strip()
    
    return email_part, fax_part, url_part
```

```python
email, fax, url = split_string_with_exceptions(university["Other details"])
```
